```python
import os
import csv
```

```python
data_dir = '/content/drive/MyDrive/PFA/videos/'
train_dir = data_dir + 'train/'
val_dir = data_dir + 'validation/'
test_dir = data_dir + 'test/'
```

```python


# Set the video folder path
def gen(video_folder,csv_path):

  # Create a list to store the data rows
  data_rows = []

  # Iterate over the video files in the folder
  for video_file in os.listdir(video_folder):
      video_path = os.path.join(video_folder, video_file)
      video_name = os.path.splitext(video_file)[0]
      components = video_name.split('_')
      video_data = [video_path] + components
      data_rows.append(video_data)

  # Define the CSV file path
  csv_file = csv_path

  # Write the data rows to the CSV file
  with open(csv_file, 'w', newline='') as file:
      writer = csv.writer(file)
      writer.writerow(['Video Path', 'F1', 'F2', 'F3', 'F4','F5'])  # Write header row
      writer.writerows(data_rows)

  print('CSV file created successfully.')
```

```python
gen(train_dir,data_dir+'video_train.csv')
```

```python
gen(val_dir,data_dir+'video_val.csv')
```

```python
gen(test_dir,data_dir+'video_test.csv')
```
