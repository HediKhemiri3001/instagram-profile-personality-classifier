```python
import pandas as pd
import numpy as np
import os
import csv
import datetime
import cv2

import seaborn as sns
import matplotlib.pyplot as plt

import re
import ast
from PIL import Image
import matplotlib.image as mpimg

df = pd.read_csv('/content/drive/MyDrive/PFA/DATASET/trainset.csv',encoding='ISO-8859-1')
df_validation = pd.read_csv('/content/drive/MyDrive/PFA/DATASET/validationset.csv',encoding='ISO-8859-1')
df_test = pd.read_csv('/content/drive/MyDrive/PFA/DATASET/testset.csv',encoding='ISO-8859-1')
```

```python
def generate_videos(df,output_video_folder):

    df=df[["user_name","Dates","Medias","Likes","Comments","I-E","S-N","T-F","J-P"]]
    df['Medias'] = df['Medias'].apply(ast.literal_eval)
    # Create the output video folder if it doesn't exist
    if not os.path.exists(output_video_folder):
        os.makedirs(output_video_folder)
        
    for _, row in df.iterrows():
        # Set up output video path with user name
        if row["I-E"] == 0 : 
            classification = "extrovert_"
        else : 
            classification = "introvert_"
        if row["S-N"] == 0 : 
            classification += "intuitive_"
        else : 
            classification += "sensing_"
        if row["T-F"] == 0 : 
            classification += "feeling_"
        else : 
            classification += "thinking_"
        if row["J-P"] == 0: 
            classification += "perceiving_"
        else : 
            classification += "judging_"
            
        output_video_path = os.path.join(output_video_folder,classification + row.user_name + '.mp4')
        
        # Set up video writer with the same frame size as the input images
        first_image = cv2.imread(row.Medias[0])
        height, width, _ = first_image.shape
        video_writer = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), 30, (width, height))
        
        # Loop through each image path and add it to the video
        for image_path in row.Medias:
            image = cv2.imread(image_path)
            video_writer.write(image)
        
        # Release video writer and print message when done
        video_writer.release()
        print(f'Video for {row.user_name} created .')
```

```python
generate_videos(df,'/content/drive/MyDrive/PFA/videos/train/')
```

```python
generate_videos(df_validation,'/content/drive/MyDrive/PFA/videos/validation/')
generate_videos(df_test,'/content/drive/MyDrive/PFA/videos/test/')
```

```python
import os
import shutil

def move_to(source_folder,destination_folder,classif):
  if not os.path.exists(destination_folder):
      os.makedirs(destination_folder)
  for filename in os.listdir(source_folder):
      if classif in filename:
          source_file = os.path.join(source_folder, filename)
          destination_file = os.path.join(destination_folder, filename)
          shutil.copy(source_file, destination_file)
```
